# Rent price in NYC

## 1. Import Modules:

```python
import numpy as np
import pandas as pd
```

## 2. Download and clean the data

### 2.1. Read the dataset into pandas

```python
df_rent = pd.read_csv('medianAskingRent_All.csv')

df_rent.head()
```

### 2.2. Calculate average rent price for each area

```python
df_rent['Average rent price'] = df_rent.mean(axis=1)
df_rent
```

```python
df_rent['Average rent price'].dtypes
```

### 2.3. Clean up and arrange the data

```python
#drop whole row with "submarket" in 'areaType' column
df_rent.drop(df_rent.index[df_rent['areaType'] == 'submarket'], inplace = True)

#reset index because we dropped rows
df_rent.reset_index(drop=True, inplace=True)

# for sake of consistency, let's also make all column labels of type string
df_rent.columns = list(map(str, df_rent.columns))

df_rent
```

```python
#set the areaName as index
df_rent.set_index('areaName', inplace=True)

#sort data descending based on mean rent price
df_rent.sort_values(['Average rent price'], ascending=False, axis=0, inplace=True)
df_rent
```

```python
#drop whole row with Nan in 'mean' column
df_rent.dropna(subset=["Average rent price"], axis=0, inplace=True)
df_rent
```

## 3. Visualise the data

```python
# use the inline backend to generate the plots within the browser
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__)
```

### 3.1. Plot average rent price in neighborhoods in a barchart

```python
#plot all areas in one barchart
df_rent['Average rent price'].plot(kind='barh', figsize=(12, 12), color='steelblue')
plt.xlabel('Average rent price')
plt.title('Average rent price in New York City neighbourhoods')

# annotate value labels to each area
for index, value in enumerate(df_rent['Average rent price']): 
    label = format(float(value)) 


plt.show()
```

```python
# get top 80 areas
df_top80 = df_rent['Average rent price'].head(80)
df_top80
```

```python
#plot top 80 areas
df_top80.plot(kind='barh', figsize=(12, 12), color='steelblue')
plt.xlabel('Average rent price')
plt.title('Average rent price in top 80 New York City neighbourhoods')

# annotate value labels to each area
for index, value in enumerate(df_top80): 
    label = format(float(value))


plt.show()
```

```python
# get bottom 80 areas
df_bottom80 = df_rent['Average rent price'].tail(80)
df_bottom80
```

```python
#plot bottom 80 areas
df_bottom80.plot(kind='barh', figsize=(12, 12), color='steelblue')
plt.xlabel('Average rent price')
plt.title('Average rent price in bottom 80 New York City neighbourhoods')

# annotate value labels to each area
for index, value in enumerate(df_bottom80): 
    label = format(float(value))


plt.show()
```

### 3.2. Plot average rent price in boroughs in a barchart

```python
#group data by boroguh
df_group_borough = df_rent[['Borough','Average rent price']]
df_group_borough = df_group_borough.groupby(['Borough'],as_index=False).mean()
df_group_borough.set_index('Borough', inplace=True)
df_group_borough.sort_values(['Average rent price'], ascending=False, axis=0, inplace=True)
df_group_borough
```

```python
#plot average rent prices by borough
df_group_borough.plot(kind='barh', figsize=(12, 12), color='steelblue')
plt.xlabel('Average rent price')
plt.title('Average rent price in New York City boroughs')

for index, value in enumerate(df_group_borough): 
    label = format(str(value))
    
plt.show()
```
